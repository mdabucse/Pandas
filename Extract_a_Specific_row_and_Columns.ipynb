The Pandas Offer 4 Major methods to Select The data Items:

1. Dot-Notation : Select a single column only it does not allows space and multiple Columns

2. loc : Based On The index and columns names

3. iloc: Based On The index and columns numbers

4. iat :select one item only based on the column and row number
