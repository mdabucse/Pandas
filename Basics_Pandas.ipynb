```python
#Libraries
import numpy as np
import pandas as pd
```

```python
#List Initialazation
arr=np.random.randint(0,100,(5,5))
```

```python
print(arr)
```

```python
#Convert Into Data table
df=pd.DataFrame(arr)
df
```

```python
#Now I Want to Change The Row And Colum Name
Row=["AIDS","AIML","ECE","CSBS","MECH"]
Col=["PSC","PYTHON","DSA","HTML","CSS"]
```

```python
df=pd.DataFrame(arr,index=Row,columns=Col)
print(df)
```

```python
#Using Dictionarys
myDic={
    'Jan':[1,2,3,4,5],
    'Feb':[5,6,7,8,9],
    'Mar':[10,11,12,13,14]
}
```

```python
print(myDic)
```

```python
df=pd.DataFrame(myDic,index=["Q","W","E","R","S"])
```

```python
print(df)
```
